```python
color=sns.color_palette()[0]
```

```python
sns.barplot(data=sorted_df, x='Category', y='Score', color=color)
plt.xticks(rotation=90);
```

```python
sorted_df.head()
```

From this barchart showing all categories, we can see our top 5 categories

They include:
1. Animals - 74965
2. Science - 71168
3. Healthy Eating - 69339
4. Technology - 68738
5. Food - 66676

To see the percentage breakdown amongst them 5 specifically; we use a pie chart

```python
piedf=sorted_df.head()
```

```python
piedf
```

```python
explode = (0.15,0.05,0.025, 0, 0)
plt.pie(labels=piedf['Category'], x=piedf['Score'], autopct='%.2f%%', startangle=0, explode=explode)
plt.title('Top 5 Most Popular Categories in their Percentage Makeup');
```

```python
sns.barplot(data=piedf, x='Category', y='Score', color=color)
plt.xlabel('Category')
plt.ylabel('Popularity')
plt.title('Popularity of Categories')
plt.xticks(rotation=15);
```
